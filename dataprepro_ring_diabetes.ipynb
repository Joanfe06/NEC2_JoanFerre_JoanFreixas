# Data Preprocessing and Plotting 
## A2-ring-merged.txt

This code snippet showcases a basic data preprocessing and visualization task using Python's data science libraries. The focus is on importing and manipulating data from the 'A2-ring-merged.txt' file and creating a scatter plot with customized aesthetics.

## Libraries Used

- **pandas (pd):** Data manipulation and reading tab-separated text files.
- **numpy (np):** Numerical operations on the data.
- **matplotlib.pyplot (plt):** Plot creation and visualization.
- **sklearn.preprocessing.MinMaxScaler:** Data scaling (not used explicitly in this snippet).

```python
# import data preprocessing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# import A2-ring-merged.txt as a dataframe
df_merged = pd.read_csv('A2-ring-merged.txt', sep='\t', header=None)

# make a plot with the first column as x-axis and the second column as y-axis and the third column as the color (0 red, 1 blue) with dots of size 10
plt.scatter(df_merged[0], df_merged[1], c=df_merged[2], s=10)

# title the plot
plt.title('A2-ring-merged.txt')

plt.show()
```

```python
df_separable = pd.read_csv('A2-ring-separable.txt', sep='\t', header=None)
plt.scatter(df_separable[0], df_separable[1], c=df_separable[2], s=10)
plt.title('A2-ring-separable.txt')
plt.show()
```

```python
df_test = pd.read_csv('A2-ring-test.txt', sep='\t', header=None)
plt.scatter(df_test[0], df_test[1], c=df_test[2], s=10)
plt.title('A2-ring-test.txt')
plt.show()
```

# Diabetes Data Preprocessing and Visualization

This code snippet illustrates the preprocessing and visualization of data from the 'diabetes.csv' file using Python's data science libraries.

```python

# import diabetes.csv as a dataframe
df_diabetes = pd.read_csv('diabetes.csv')

# put column names in a list
col_names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction',
             'Age']

# remove rows with 0 as values for columns Glucose, BloodPressure, SkinThickness, Insulin, BMI
df_diabetes = df_diabetes[(df_diabetes['Glucose'] != 0) & (df_diabetes['BloodPressure'] != 0) & (
        df_diabetes['SkinThickness'] != 0) & (df_diabetes['Insulin'] != 0) & (df_diabetes['BMI'] != 0)]

# normalize the dataframe
scaler = MinMaxScaler()
df_diabetes[col_names] = scaler.fit_transform(df_diabetes[col_names])

print(df_diabetes.head())

#print number of rows and columns
print(df_diabetes.shape)

# plot each column of the dataframe with the index as x-axis and the column values as y-axis and the outcome as the color (0 red, 1 blue)
for i in col_names:
    plt.scatter(df_diabetes.index, df_diabetes[i], c=df_diabetes['Outcome'], s=10)
    plt.title('diabetes.csv')
    plt.xlabel('index')
    plt.ylabel(i)
    plt.show()
```
